## **Feature Engineering**

```python
# add a col to indicate the earthquake center
df_17 = df_train[df_train.geo_level_1_id==17].groupby('geo_level_2_id').mean()
list_17_2 = df_17[df_17.damage_grade>2.5].index # the geo_2_id that has high damage_grade when geo_1_id=17
X_train['center'] = np.where((df_train.geo_level_1_id==17)&(df_train.geo_level_2_id.isin(list_17_2)), 1, 0)

# add the indicator to validation data
df_val = pd.concat([X_val, y_val], axis=1)
df_17 = df_val[df_val.geo_level_1_id==17].groupby('geo_level_2_id').mean()
list_17_2 = df_17[df_17.damage_grade>2.5].index # the geo_2_id that has high damage_grade when geo_1_id=17
X_val['center'] = np.where((df_val.geo_level_1_id==17)&(df_val.geo_level_2_id.isin(list_17_2)), 1, 0)
```

## **Make Pipeline**

```python
def make_pipeline(clf=None):
    """
    use pipeline combine preprocessing and classifier training together
    """
    numericals = ["geo_level_1_id", "geo_level_2_id", "geo_level_3_id",
                  "count_floors_pre_eq", "age", "area_percentage",
                  "height_percentage"]

    categoricals = ["land_surface_condition", "foundation_type",
                    "roof_type", "ground_floor_type", "other_floor_type",
                    "position", "plan_configuration"]
    
    # for categorical features, dummy it  
    categorical_features = categoricals
    categorical_transformer = Pipeline(steps=[
        ('onehot', preprocessing.OneHotEncoder())])
    
    if type(clf) in [SVC,MLPClassifier]:
        numeric_features = numericals 
    # for numeric features, normalize it 
        numeric_transformer = Pipeline(steps=[
        ('scaler', preprocessing.StandardScaler())])

        preprocessor = compose.ColumnTransformer(
            transformers=[
                ('num', numeric_transformer, numeric_features),
                ('cat', categorical_transformer, categorical_features)])

    else: # for randomforest and lgb, we don't need normalize numeric features 
        preprocessor = compose.ColumnTransformer(transformers=[('cat', categorical_transformer, categorical_features)])
        
    pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                          ('classifier',clf)])
    
    return pipeline
```

## **Model 1: SVM**

svm with kernel=poly

```python
# try SVM, since it's imbalanced dataset 
svm = SVC(kernel='poly', degree=3, gamma='auto', decision_function_shape='ovo', max_iter=500)
pipeline_svm = make_pipeline(svm)

pipeline_svm.fit(X_train, y_train)
```

```python
y_train_pred_svm = pipeline_svm.predict(X_train)
train_score_svm = f1_score(y_train, y_train_pred_svm, average='micro')
print(f"SVM_poly train f1 score: {round(train_score_svm, 4)}")

# micro-average is preferable because there is class imbalance
y_val_pred_svm = pipeline_svm.predict(X_val)
val_score_svm = f1_score(y_val, y_val_pred_svm, average='micro')

print(f"SVM_poly validation f1 score: {round(val_score_svm, 4)}")
# in stead of digging deep into SVM parameters, we'd like to do with RF/LGB, which is more promising
```

```python
# we can also look at the other metrics for this classfication method 
report_svm = classification_report(y_val, y_val_pred_svm,output_dict=True,digits=4)
print('-- Validation Set Metric Report For SVM -- \n')
print(f"-- Main metric -- \n\nmicro-weighted f1 score: {round(val_score_svm, 4)}\n")
print("-- Other metric -- \n")

for key, value in report_svm.items():
    print(key, value)
```

## **Model 3: Random Forest**

```python
# seperate different type of features 
numericals = ["geo_level_1_id", "geo_level_2_id", "geo_level_3_id",
              "count_floors_pre_eq", "age", "area_percentage",
              "height_percentage"]

categoricals = ["land_surface_condition", "foundation_type",
                "roof_type", "ground_floor_type", "other_floor_type",
                "position", "plan_configuration"]

binaries = ['has_secondary_use', 'center']

def data_prep(X_train, X_val):
    """
    data processing for randomforest 
    one-hot coding for categorical variables
    """
    cols_to_include = numericals + binaries + categoricals
    X_train, X_val = X_train[cols_to_include], X_val[cols_to_include]
    
    for cat in categoricals:
        # one-hot encoding
        one_hot = preprocessing.OneHotEncoder(handle_unknown="ignore")
        hots_array = one_hot.fit_transform(X_train[[cat]]).toarray()
        columns = [cat + f"_{i}" for i, _ in enumerate(one_hot.get_feature_names())] # new col names
        hots_df = pd.DataFrame(hots_array, columns=columns, index=X_train.index)
        X_train = pd.concat([X_train, hots_df], axis=1).drop(columns=[cat])
        
        # now for validation
        hots_array = one_hot.transform(X_val[[cat]]).toarray()
        hots_df = pd.DataFrame(hots_array, columns=columns, index=X_val.index)
        X_val = pd.concat([X_val, hots_df], axis=1).drop(columns=[cat])
        
    return X_train, X_val

X_train, X_val = data_prep(X_train, X_val)
```

```python
def rf_make_random_cv():
    """
    use random search to choose best parameters for randomforest model
    """
    rf_hyperparameters = dict(n_estimators=np.arange(60,70,5), 
                          max_features=np.arange(15,17,1),  
                          min_samples_leaf=np.arange(3,5,1))
    
    rf = RandomForestClassifier(random_state=42)
    
    rf_random_cv = RandomizedSearchCV(rf,  
                              rf_hyperparameters, 
                              scoring = 'f1_micro',
                              cv = 3,
                              n_iter=10,
                              verbose=0)   
    return rf_random_cv

rf_random = rf_make_random_cv()
# it takes 10min to run 
rf_random.fit(X_train, y_train)
```

```python
# look at the best estimators found by random search 
rf_random.best_estimator_
```

```python
# look at the training and validation micro-weighted F1 score for RF
y_train_pred_rf = rf_random.predict(X_train)
train_score_rf = f1_score(y_train, y_train_pred_rf, average='micro')

print(f"RandomForest train f1 score: {round(train_score_rf, 4)}")
y_val_pred_rf = rf_random.predict(X_val)
val_score_rf = f1_score(y_val, y_val_pred_rf, average='micro')

print(f"RandomForest validation f1 score: {round(val_score_rf, 4)}")
```

```python
# we can also look at the other metrics for this classfication method 
report_rf = classification_report(y_val, y_val_pred_rf,output_dict=True,digits=4)
print('-- Validation Set Metric Report For RF -- \n')
print(f"-- Main metric -- \n\nmicro-weighted f1 score: {round(val_score_rf, 4)}\n")
print("-- Other metric -- \n")

for key, value in report_rf.items():
    print(key, value)
```

## **Model 4: Gradient Boosting**

**Model Training**

```python
"""
tunning gbm model parameters and train model
"""
d_train = lgb.Dataset(X_train, label=y_train-1)
params = {}
params['learning_rate'] = 0.0028
params['num_class']=3
params['boosting_type'] = 'gbdt'
params['objective'] = 'multiclass'
params['metric'] = 'multi_logloss'
params['sub_feature'] = 0.8
params['num_leaves'] = 5000
clf = lgb.train(params, d_train, 100)
```

**More about confusion_matrix:**

```python
def cm_analysis(y_true, y_pred, labels, filename=None, ymap=None, figsize=(10,10)):
    """
    Code obtained from: https://gist.github.com/hitvoice/36cf44689065ca9b927431546381a3f7
    
    Generate matrix plot of confusion matrix with pretty annotations.
    The plot image is saved to disk.
    args: 
      y_true:    true label of the data, with shape (nsamples,)
      y_pred:    prediction of the data, with shape (nsamples,)
      filename:  filename of figure file to save
      labels:    string array, name the order of class labels in the confusion matrix.
                 use `clf.classes_` if using scikit-learn models.
                 with shape (nclass,).
      ymap:      dict: any -> string, length == nclass.
                 if not None, map the labels & ys to more understandable strings.
                 Caution: original y_true, y_pred and labels must align.
      figsize:   the size of the figure plotted.
    """
    if ymap is not None:
        y_pred = [ymap[yi] for yi in y_pred]
        y_true = [ymap[yi] for yi in y_true]
        labels = [ymap[yi] for yi in labels]
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    cm_sum = np.sum(cm, axis=1, keepdims=True)
    cm_perc = cm / cm_sum.astype(float) * 100
    annot = np.empty_like(cm).astype(str)
    nrows, ncols = cm.shape
    for i in range(nrows):
        for j in range(ncols):
            c = cm[i, j]
            p = cm_perc[i, j]
            if i == j:
                s = cm_sum[i]
                annot[i, j] = '%.1f%%\n%d/%d' % (p, c, s)
            elif c == 0:
                annot[i, j] = ''
            else:
                annot[i, j] = '%.1f%%\n%d' % (p, c)
    cm = pd.DataFrame(cm, index=labels, columns=labels)
    cm.index.name = 'Actual'
    cm.columns.name = 'Predicted'
    fig, ax = plt.subplots(figsize=figsize)
    sns.heatmap(cm, annot=annot, fmt='', ax=ax, cmap=sns.color_palette("Blues"))
    b, t = plt.ylim() # discover the values for bottom and top
    b += 0.5 # Add 0.5 to the bottom
    t -= 0.5 # Subtract 0.5 from the top
    plt.ylim(b, t) # update the ylim(bottom, top) values
    plt.show() # ta-da!
    if filename:
        plt.savefig(filename)
```

```python
cm_analysis(y_val, new, 
            labels=[1,2,3], ymap={1:"low damage", 2:"medium damage", 3:"complete destruction"},
           figsize=(8,8))
```

**SMOTE oversampling**



```python
sm = SMOTE()
resampled_training_inputs,\
resampled_training_outputs_labels = sm.fit_resample(X_train, y_train)
```

```python
d_train = lgb.Dataset(resampled_training_inputs,\
                      label=resampled_training_outputs_labels-1)

params = {}
params['learning_rate'] = 0.0028
params['num_class']=3
params['boosting_type'] = 'gbdt'
params['objective'] = 'multiclass'
params['metric'] = 'multi_logloss'
params['sub_feature'] = 0.8
params['num_leaves'] = 5000

clfs = lgb.train(params, d_train, 100)
```

```python
ly_predict = clfs.predict(X_train)
new = np.zeros(len(ly_predict))
for index,i in enumerate(ly_predict):
    new[index] = np.argmax(i)+1
train_score = f1_score(y_train, new, average='micro')
print(f"smote train f1 score: {round(train_score, 4)}")

ly_va = clfs.predict(X_val)
new = np.zeros(len(ly_va))
for index,i in enumerate(ly_va):
    new[index] = np.argmax(i)+1
val_score = f1_score(y_val, new, average='micro')
print(f"smote validation f1 score: {round(val_score, 4)}")
```

# Summary & Takeaways
- We solved this multi-classification problem on predicting building's damage-grade, and we ranked top 10% in the Drivendata Competition. 

- We did feature engineering to set a binary column indicating whether the buiding is in the  “accident  center”, and it's very useful in exploiting the geographic id (location) info. 

- We implemented 4 ML algorithms: 
    * Support Vector Machine (SVM)
    * Multi-layer Perceptron Classifier (MLP)
    * Rrandom Forest (RF)
    * Gradient Boosting Machine (GBM)
    
- We used pipeline to do the work, and tuned parameters for RF and GBM. 
- In the GBM model, we used two methods to solve the imbalanced-dataset problem:
    * SMOTE Oversample
    * Tuning Probability Threshold
- We calculated multiple evaluation metrics, like recall, precision, and f1-score. Since this dataset is imbalanced, we use micro-weighted f1-score as our main metric to evaluate our models. The final metrics for these models are:
    * GBM : 0.74 (micro-weighted f1-score)
    * RF  : 0.73 (micro-weighted f1-score)
    * MLP : 0.69 (micro-weighted f1-score)
    * SVM : 0.33 (micro-weighted f1-score)
- According to the micro-weighted f1-score, GBM is our best model in this case. We further intepretated GBM model with confusion_matrix, feature-importance plot and shap.
- The RF and GBM models worked well in this project, and they can also be expended to the building-damage prediction in other natural disasters, like flood and wild fire.
