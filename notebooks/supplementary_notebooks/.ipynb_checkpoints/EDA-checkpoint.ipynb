```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import featexp
```

```python
train_values = pd.read_csv("../data/train_values.csv")
train_labels = pd.read_csv("../data/train_labels.csv")
test_values = pd.read_csv("../data/test_values.csv")
```

```python
train_labels["damage_grade"].hist()
```

```python
## train test split, maintaining class distributions
df = pd.merge(train_values, train_labels, how="inner", on="building_id")

# first, getting all rows from each class
df_1 = df[df['damage_grade']==1]
df_2 = df[df['damage_grade']==2]
df_3 = df[df['damage_grade']==3]

# now getting splits for each
df_train_1, df_test_1 = train_test_split(df_1, test_size=0.1)
df_train_2, df_test_2 = train_test_split(df_2, test_size=0.1)
df_train_3, df_test_3 = train_test_split(df_3, test_size=0.1)

# now gluing it all back together
df_train = pd.concat([df_train_1, df_train_2, df_train_3], axis=0).sample(frac=1)
df_test = pd.concat([df_test_1, df_test_2, df_test_3],axis=0).sample(frac=1)

# finally, splitting things up into values and targets
X_train = df_train.drop(columns=["damage_grade"])
y_train = df_train["damage_grade"]
X_val = df_test.drop(columns=["damage_grade"])
y_val = df_test["damage_grade"]
```

# Idea
It seems to me that if we could get an understanding of what the `geo_level_i_id` column means, than we could get an understanding of how far away each building was from ground zero..

```python
def fix_na(df, colname):
    df[colname+'_na'] = pd.isnull(df[colname])
    df[colname].fillna(df[colname].median(), inplace=True)
```

```python
def dummy(df, col):
    dummies = pd.get_dummies(df[col])
    dummies.columns = col + "_" + dummies.columns
    df = pd.concat([df, dummies], axis=1)
    return df.drop(columns=col)
```

```python
numericals = ["geo_level_1_id", "geo_level_2_id", "geo_level_3_id",
              "count_floors_pre_eq", "age", "area_percentage",
              "height_percentage", "count_families"]

categoricals = ["land_surface_condition", "foundation_type",
                "roof_type", "ground_floor_type", "other_floor_type",
                "position", "plan_configuration", 'legal_ownership_status']

binaries = ['has_superstructure_adobe_mud',
       'has_superstructure_mud_mortar_stone', 'has_superstructure_stone_flag',
       'has_superstructure_cement_mortar_stone',
       'has_superstructure_mud_mortar_brick',
       'has_superstructure_cement_mortar_brick', 'has_superstructure_timber',
       'has_superstructure_bamboo', 'has_superstructure_rc_non_engineered',
       'has_superstructure_rc_engineered', 'has_superstructure_other',
         'has_secondary_use',
       'has_secondary_use_agriculture', 'has_secondary_use_hotel',
       'has_secondary_use_rental', 'has_secondary_use_institution',
       'has_secondary_use_school', 'has_secondary_use_industry',
       'has_secondary_use_health_post', 'has_secondary_use_gov_office',
       'has_secondary_use_use_police', 'has_secondary_use_other']
```

```python
def data_prep(df):
    df = df[numericals + categoricals + binaries]
    for cat in categoricals:
        df = dummy(df, cat)
    return df
```

# Initial go at modeling

For some categoricals, maybe try feature encoding, instead of one-hot encoding?

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
```

```python
X_train = data_prep(X_train)
X_val = data_prep(X_val)
```

```python
rf = RandomForestClassifier(n_estimators=20, min_samples_leaf=3, random_state=742)
```

```python
rf.fit(X_train, y_train)
```

```python
y_train_pred = rf.predict(X_train)
train_score = f1_score(y_train, y_train_pred, average='micro')
print(f"train f1 score: {round(train_score, 4)}")
```

```python
y_val_pred = rf.predict(X_val)
val_score = f1_score(y_val, y_val_pred, average='micro')
print(f"validation f1 score: {round(val_score, 4)}")
```
